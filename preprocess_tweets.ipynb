```python
# Intital function to clean basic elements of tweets
def clean_tweet(tw):
    tw = tw.strip()
    
    # Remove links
    tw = re.sub(r'https?:\/\/.*[\r\n]*', '', tw, flags=re.MULTILINE)
    
    # Remove "RT" tags
    tw = re.sub('RT ', '', tw)
#     tw = re.sub(r"RT @\S+", "", tw)

    # Remove Mentions
    tw = re.sub(r"@\S+", "", tw)
    
    # Replace "\n" by space
    tw = tw.replace('\n', ' ')
    
    return tw.strip()
```

```python
clean_p_list = [clean_tweet(tw) for tw in p_list]
```

```python
import emoji
```

```python
# Function to detect and extract emojis from the text
def extract_emojis(str):
    return ''.join(c for c in str if c in emoji.UNICODE_EMOJI)
```

```python
import string

# Function to remove the emojis and punctuations from the dataframe

def clean_text(tw):
    allchars = [str for str in tw]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    tw = ' '.join([str for str in tw.split() if not any(i in str for i in emoji_list)])
    
    tw= re.sub(r"\n", " ", tw)
    
    tw = tw.translate("#-:()[]/*~+|'")

    return tw.strip()
```

```python
final_clean_df = pd.DataFrame()
final_clean_df['text'] = final_processed_clean_df['text'].apply(clean_text)
final_clean_df['emoji'] = final_processed_clean_df['emoji']
```

```python
# Use a Twitter-specialized tokenizers to create dicts 
from nltk.tokenize import TweetTokenizer

tknzr = TweetTokenizer()
```

```python
import collections
```

```python
# Tokenize tweets and count out <unk> words
vocab = collections.Counter()
i = 0
for tw in final_clean_df['text']:
    for word in tknzr.tokenize(tw):
        vocab[word]+=1
```

```python
# look at vocab
vocab.most_common(50)
```

```python
#stats regarding the vocab
print("Num of words having count >=50:", len([count for x, count in vocab.items() if count >= 50]))
print("Num of words having count >=10:", len([count for x, count in vocab.items() if count >= 10]))
print("Num of words having count >=5:", len([count for x, count in vocab.items() if count >= 5]))

# percentage of unks
print("Precentage of UNKS count < 5:", sum([count for x, count in vocab.items() if count < 5]) * 1.0/sum([count for x, count in vocab.items()]))
```

```python
# Convert rare words to unk. rare words are converted to <UNK>

THRESHOLD = 5
unk_word = set([w for w, count in vocab.items() if count < 5])

total_words = 0
unk = 0

tweets_tokenized = []

for tw in final_clean_df['text']:
    new = []
    for word in tknzr.tokenize(tw):
        if '<E' in word or word not in unk_word:
            new.append(word)
        else:
            new.append('<UNK>')
    tweets_tokenized.append(' '.join(new))
```
